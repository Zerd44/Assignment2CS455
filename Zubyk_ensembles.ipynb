```python
from sklearn.datasets import fetch_openml
import numpy as np
```

Import the needed information

```python
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
```

```python
mnist.keys()
```

```python
mnist.data.shape
```

```python
mnist.target.shape
```

Split the model for test and train

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=42)
for train_index, test_index in split.split(mnist.data, mnist.target):
    X_train, X_test = mnist.data[train_index], mnist.data[test_index]
    y_train, y_test = mnist.target[train_index], mnist.target[test_index]
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
dt_pred = dt.predict(X_test)
print("Predictions for dt ", list(dt_pred))
print("Labels for dt ", list(y_test))
```

```python
dt_accuracy = dt.score(X_test, y_test)
dt_accuracy
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
rf_pred = rf.predict(X_test)
print("Predictions for rf ", list(rf_pred))
print("Labels for rf ", list(y_test))
```

```python
rf_accuracy = rf.score(X_test, y_test)
rf_accuracy
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
knn_pred = knn.predict(X_test)
print("Predictions for knn ", list(knn_pred))
print("Labels for knn ", list(y_test))
```

```python
knn_accuracy = knn.score(X_test, y_test)
knn_accuracy
```

Voting Classifier

```python
from sklearn.ensemble import VotingClassifier
```

Hard Voting

```python
hv = VotingClassifier(estimators=[('dt', dt), ('rf',rf),('knn',knn)], voting='hard')
hv.fit(X_train, y_train)
```

```python
hv_pred = hv.predict(X_test)
print("Predictions for hv ", list(hv_pred))
print("Labels for hv ", list(y_test))
```

```python
hv_accuracy = hv.score(X_test, y_test)
hv_accuracy
```

Soft Voting

```python
sv = VotingClassifier(estimators=[('dt', dt), ('rf',rf),('knn',knn)], voting='soft')
sv.fit(X_train, y_train)
```

```python
sv_pred = sv.predict(X_test)
print("Predictions for sv ", list(sv_pred))
print("Labels for sv ", list(y_test))
```

```python
sv_accuracy = sv.score(X_test, y_test)
sv_accuracy
```

Decision Tree with bagging

```python
from sklearn.ensemble import BaggingClassifier
```

```python
dt_bagging = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=10, bootstrap=True, n_jobs=-1)
dt_bagging.fit(X_train, y_train)
```

```python
dt_bagging_pred = dt_bagging.predict(X_test)
print("Predictions for dt_bagging ", list(dt_bagging_pred))
print("Labels for dt_bagging ", list(y_test))
```

```python
dt_bagging_accuracy = dt_bagging.score(X_test,y_test)
dt_bagging_accuracy
```

Decision tree with Pasting

```python
dt_pasting = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=10, bootstrap=False, n_jobs=-1)
dt_pasting.fit(X_train, y_train)
```

```python
dt_pasting_pred = dt_pasting.predict(X_test)
print("Predictions for dt_pasting ", list(dt_pasting_pred))
print("Labels for dt_pasting ", list(y_test))
```

```python
dt_pasting_accuracy = dt_pasting.score(X_test, y_test)
dt_pasting_accuracy
```

AdaBoost classifier

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ab = AdaBoostClassifier(n_estimators=50, learning_rate=1.0)
ab.fit(X_train, y_train)
```

```python
ab_pred = ab.predict(X_test)
print("Predictions for AdaBoost ", list(ab_pred))
print("Labels for AdaBoost ", list(y_test))
```

```python
ab_accuracy = ab.score(X_test, y_test)
ab_accuracy
```
